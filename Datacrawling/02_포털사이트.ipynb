### 네이트 사이트의 메뉴 추출
- 메뉴명과 연결 url을 추출

```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
url = 'https://www.nate.com'
url
```

```python
html = urlopen(url)
html
type(html)
```

```python
html_text = html.read() # 전달된 소스코드만 추출
```

```python
# bs4 객체 바이너리코드를 집어넣어도 bs_obj에 의해 변환된다.
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')
```

```python
print(bs_obj.prettify())
```

```python
#divGnb
menu = bs_obj.find('div', {'id':'divGnb'})
menu_li = menu.select('li')
# a태그를 가지고 오기위해 menu_li를 만져야함. 근데 이게 리스트기 때문에 find안됨.
menu_li
```

```python
# 리스트는 이렇게 처리해야됨
menu_li[0]
menu_li[0].text
menu_li[0].a
menu_li[0].a['href']
```

```python
for li in menu_li :
    print(li.text,li.a['href'])
```

### 네이트 메뉴명과 링크 url df로 생성 후 저장
- 빈 df를 생성 후 df에 data를 행으로 추가
- 각각의 행을 df로 생성 후 concat 함수로 결합

```python
# 빈 df 생성
nate_menu = pd.DataFrame({'메뉴':[], '링크':[]})
nate_menu
```

```python
index = 0
for li in menu_li :
    menu_text = li.text
    menu_link = li.a['href']
    if menu_link == '' or menu_link == 'javascript:;' or menu_text in nate_menu['메뉴'].values :
        continue # 아래 코드 실행하지 않음
    temp = pd.DataFrame({'메뉴':menu_text,'링크':menu_link},index=[index])
    nate_menu = pd.concat([nate_menu, temp])
    index = index +1
```

```python
'뉴스' in nate_menu['메뉴']
'뉴스' in nate_menu['메뉴'].values
```

```python
# 네이버 뉴스 메인 페이지에서 뉴스섹션명과 링크 url 추출
url = 'https://news.naver.com'
html = urlopen(url)
```

### 네이버 뉴스 섹션 메뉴의 링크 df로 구성
- 리스트에 각 컬럼 데이터를 모은 후 df로 구성

```python
section = []
link = []
```

```python
for li in lis : 
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict = {'section':section,'link':link}
naver_news_df = pd.DataFrame(col_dict)
```

### 개별연습

```python
# 모든 과정에 왜그런지 상세하게 정확하게 알아야 한다.
url = 'https://news.naver.com/'
html = urlopen(url) # 브라우저를 열어 url을 넣고 주소로 들어가는 과정을 해주는 것이다. 
# 브라우저가 클라이언트로서 해당 url주소에 응답요청을 한다. 요청을 하고 응답신호와 data를 얻어온다.
html
#HTTPResponse의 의미는 요청에 대한 응답신호를 가지고있는 객체이다. HTTP신호로 요청을 하고 HTTP 응답을 반아온 객체이다.
html_text = html.read() # 언어온 데이터중 소스코드를 얻어와야 한다. 디코딩 되지 않은 바이너리 코드로 불러와진다. 한번 읽어오면 객체변수 안의 소스코드가 삭제됨.
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')#필요한 데이터를 추출하기 위해 파싱객체로 만들어야 한다.
#bs4 가 가지고 있는 BeautifulSoup함수를 통해 파싱객체로 만든다. 파싱하려는 데이터를 html문법으로 파싱하겠다는 뜻이다.
type(bs_obj) #bs4.BeautifulSoup는 bs4의 함수를 이용할 수 있는 상태가 된 것이다.
```

## bs4객체가 저장하고 있는 소스코드에서 지정한 태그를 찾아오는 함수
- find(태그명 {'속성명':'속성값'})
- find_all/findAll(태그명, {'속성명':'속성값'})
- select('css selector')
- 보고있는 결과와 다른 이유는 처음 데이터가 넘어온 시점이랑 내가 지금 보고있는 순간의 암튼다름

```python
aticle = bs_obj.findAll('div', {'class':'cjs_nf_list _item_list'})
title = aticle[0].findAll('h4')
name = aticle[0].findAll('div', {'class':'cn_name'})

title_list = []
name_list = []
for _ in title:
    title_list.append(_.text)
for _ in name:
    name_list.append(_.text)

print(list(zip(title_list, name_list)))
```
